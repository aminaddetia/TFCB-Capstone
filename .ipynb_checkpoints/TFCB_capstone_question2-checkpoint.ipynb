```python
import pandas as pd
import matplotlib.pyplot as plt

ds_train = pd.read_csv('data/ds_train.csv')
ds_test = pd.read_csv('data/ds_test.csv')

ds_train['Set'] = 'r'

ds_test['Set'] = 'b'

ds_full = pd.concat([ds_train, ds_test])

ds_full = ds_full.reset_index()

ds_pca_input = ds_full.drop(['index', 'RL', 'Set'], axis=1)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=8)

pca.fit(ds_pca_input)

ds_pca = pca.transform(ds_pca_input)

plt.scatter(ds_pca[:,0], ds_pca[:,1], c = ds_full['Set'])
```
